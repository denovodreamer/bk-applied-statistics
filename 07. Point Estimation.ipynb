```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 40, 0.001)

y = stats.lognorm.pdf(x, s=2, loc=0, scale=0.75)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue')
```

```python
def lognormal_sample(n):
    return np.random.lognormal(mean=2, sigma=0.75, size=n)

def normal_sample(n):
    return np.random.normal(loc=2, scale=0.75/np.sqrt(n), size=n)
```

```python
lognormal_sample(10).reshape(1, -1)
```

```python
samples = np.concatenate([lognormal_sample(10).reshape(1, -1) for _ in range(10)])
```

```python


import numpy as np
import matplotlib.pyplot as plt

# Example numpy array
data = lognormal_sample(10000)  # generating random data

# Plotting the histogram
plt.hist(data, bins=35, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
samples = np.concatenate([lognormal_sample(10000).reshape(1, -1) for _ in range(1000)])

sample_means = samples.mean(axis=1)

plt.hist(sample_means, bins=35, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example numpy array
data = sample_means  # generating random data from a normal distribution

# Sort the sample data
sorted_data = np.sort(data)

# Compute the theoretical quantiles
n = len(sorted_data)
theoretical_quantiles = np.linspace(0, 1, n+2)[1:-1]  # evenly spaced probabilities, excluding 0 and 1
theoretical_quantiles = np.quantile(np.random.normal(np.mean(data), np.std(data), 1000000), theoretical_quantiles)

# Plotting the Q-Q plot
plt.scatter(theoretical_quantiles, sorted_data)
plt.plot(theoretical_quantiles, theoretical_quantiles, color='r', linestyle='--')  # 45-degree line for reference
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

```python


def exp_sample(n):
    lam = 0.5
    return np.random.exponential(scale=1/lam, size=n)


samples = np.concatenate([exp_sample(100).reshape(1, -1) for _ in range(10000)])

sample_means = samples.mean(axis=1)

plt.hist(sample_means, bins=35, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
