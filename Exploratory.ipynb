```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate data for different distributions
normal_data = np.random.normal(loc=0, scale=1, size=1000)
exponential_data = np.random.exponential(scale=1, size=1000)
negative_exponential_data = -np.random.exponential(scale=1, size=1000)
uniform_data = np.random.uniform(low=-1, high=1, size=1000)

# Create a figure and axis for the box plots
fig, ax = plt.subplots(1, 4, figsize=(16, 6))

# Plot box plots for each distribution
ax[0].boxplot(normal_data, vert=False)
ax[0].set_title('Normal Distribution')

ax[1].boxplot(exponential_data, vert=False)
ax[1].set_title('Right-Skewed Distribution')

ax[2].boxplot(negative_exponential_data, vert=False)
ax[2].set_title('Left-Skewed Distribution')

ax[3].boxplot(uniform_data, vert=False)
ax[3].set_title('Uniform Distribution')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate data for normal distributions with the same mean but different standard deviations
std_dev_1_data = np.random.normal(loc=0, scale=1, size=1000)
std_dev_2_data = np.random.normal(loc=0, scale=2, size=1000)
std_dev_3_data = np.random.normal(loc=0, scale=3, size=1000)
std_dev_0_5_data = np.random.normal(loc=0, scale=0.5, size=1000)

# Create a figure and axis for the box plots
fig, ax = plt.subplots(1, 4, figsize=(16, 6))

# Define the same x-axis limits for all plots
x_limits = [-10, 10]

# Plot box plots for each normal distribution
ax[0].boxplot(std_dev_1_data, vert=False)
ax[0].set_title('Normal Distribution (sd=1)')
ax[0].set_xlim(x_limits)

ax[1].boxplot(std_dev_2_data, vert=False)
ax[1].set_title('Normal Distribution (sd=2)')
ax[1].set_xlim(x_limits)

ax[2].boxplot(std_dev_3_data, vert=False)
ax[2].set_title('Normal Distribution (sd=3)')
ax[2].set_xlim(x_limits)

ax[3].boxplot(std_dev_0_5_data, vert=False)
ax[3].set_title('Normal Distribution (sd=0.5)')
ax[3].set_xlim(x_limits)

# Display the plots
plt.tight_layout()
plt.show()
```
