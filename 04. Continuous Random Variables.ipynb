# Probability Density Functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set seed for reproducibility
np.random.seed(42)

# Generate a dataset of 500 tensile strength measurements (mean=463.3, std=7.56)
data = np.random.normal(loc=463.3, scale=7.56, size=500)

# Calculate mean and standard deviation
mu = np.mean(data)
sigma = np.std(data, ddof=1)

# Fit a normal distribution
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution PDF')
plt.scatter(data, np.zeros_like(data), color='red', alpha=0.6, label='Data Points')
plt.axvline(mu, color='blue', linestyle='dashed', linewidth=1, label=f'Mean = {mu:.2f}')
plt.title('Tensile Strength Distribution (500 Data Points)')
plt.xlabel('Tensile Strength (MPa)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Normal Approximation

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters of the binomial distribution
n = 200  # number of trials
p = 0.4  # probability of success

# Compute the mean and standard deviation of the binomial distribution
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Define the range of x values (number of successes)
x = np.arange(0, n+1)

# Compute the binomial probabilities
binomial_probs = stats.binom.pmf(x, n, p)

# Compute the normal approximation
normal_approx = stats.norm.pdf(x, mu, sigma)

# Plot the binomial probabilities and the normal approximation
plt.figure(figsize=(10, 6))
plt.plot(x, binomial_probs, 'bo', label='Binomial PMF')
plt.plot(x, normal_approx, 'r-', label='Normal Approximation')
plt.title('Binomial Distribution and Normal Approximation')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
